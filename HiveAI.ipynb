# Applying the 4 benchmark(s) / cases to the raw data

```python
one = df.loc[(df['definition_list']==True) & ((df['semantic']==False) | (df['tabular']==False))]
two = df.loc[(df['semantic']==True) & (df['tabular']==False)]
three= df.loc[(df['tabular']==False) & ((df['semantic']==True) | (df['definition_list']==True))]
four =  df.loc[((df['header_row']==True) | (df['header_column']==True)) & (df['semantic']==False)]
df_full = pd.concat([one, two, three, four])
```

```python
merged = df_full.drop_duplicates()
```
